```python
import pandas as pd
df = pd.read_csv("../data.csv")
```

### Récupérer certaines partie du fichier avec les méthodes head() et tail()

```python
# La méthode head() va afficher les cinq première entrées

df.head()
```

```python
# la méthode tail() affiche les cinq dernières

df.tail()
```

_Ci-dessus, c'est ce qu'on fait généralement pour ne pas avoir trop d'éléments affichés. Quand on veut juste avoir un aperçu de notre data frame, on peut utiliser __Head()__ et __Tail()__._

Si jamais on souhaite afficher un peu plus d'informations que juste les cinq premiers ou les cinq derniers éléments, on peut spécifier dans ___Head()___ ou dans ___Tail()___ le nombre d'éléments qu'on souhaite afficher de la manière suivante :

```python
# Affiche les dix premières entrées

df.head(10)
```
